__K-means__

```python
from sklearn.cluster import KMeans
from collections import Counter
```

```python
for i in df.iloc[:,22:25]:
    for j in range(50):
        print('\n'+i+', '+str(j+1)+' , result:\n')
        kmeans=KMeans(n_clusters=j+1)
        kmeans.fit(df.iloc[:,:22])
        clabel=i+'_cluster'
        df[clabel]=kmeans.labels_
        for k in range(j+1):
            result=Counter(df.loc[df[clabel]==k,i]).most_common(1)[0][0]
            print('cluster '+str(k)+' most common:'+result)
```

```python
Hamm=pd.DataFrame()
for j in range(50):
    for i in df.iloc[:,22:25]:
        kmeans=KMeans(n_clusters=j+1)
        kmeans.fit(df.iloc[:,:22])
        clabel=i+'_cluster'
        df[clabel]=kmeans.labels_
        arr_clus='K_'+i
        for k in range(j+1):
            result=Counter(df.loc[df[clabel]==k,i]).most_common(1)[0][0]
            df.loc[df[clabel]==k,arr_clus]=result
    df['Hamming_distance']=(df['Family']!=df['K_Family']).astype('int')+(df['Genus']!=df['K_Genus']).astype('int')+(df['Species']!=df['K_Species']).astype('int')
    Hamm.loc[j,'avg_distance']=np.mean(df['Hamming_distance'])
    Hamm.loc[j,'Hamm_loss']=np.sum(df['Hamming_distance'])/(3*len(df))
    Hamm.loc[j,'Hamm_score']=np.sum((3-df['Hamming_distance'])/(3+df['Hamming_distance']))/len(df)
```
